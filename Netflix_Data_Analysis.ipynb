#Analyze Top 5 Actors on Netflix

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('/content/Netflix Dataset.csv')
df['Cast'] = df['Cast'].fillna('Not Cast Specified')
cast_df = pd.DataFrame()
cast_df = df['Cast'].str.split(',', expand=True).stack()
cast_df =cast_df.to_frame()
cast_df.columns = ['Actors']
actors =cast_df.groupby(['Actors']).size().reset_index(name='Total Counts')
actors = actors[actors.Actors != 'Not Cast Specified']
actors = actors.sort_values('Total Counts', ascending=False)
actorsTop5 = actors.head()
actorsTop5 = actorsTop5.sort_values('Total Counts')
barChart2=px.bar(actorsTop5, x='Total Counts', y='Actors', title='Top 5 Actors on Netflix')
barChart2.show()
```

#Analyze the Content Produced on  Netflix Based on Years

```python
df1 = df[['Type', 'Release_Date']].copy()
df1['Release_Date'] = df1['Release_Date'].str.strip()
df1['Release Year'] = pd.to_datetime(df1['Release_Date'], format='%B %d, %Y', errors='coerce').dt.year
df2 = df1.groupby(['Release Year', 'Type']).size().reset_index(name='Total Content')
df2 = df2[df2['Release Year'] >= 2010]
df2 = df2.sort_values('Release Year')
print(df2)
```

```python
graph = px.line(df2, x='Release Year', y='Total Content', color='Type', title=' Trend of Content Produced on Netflix Every Years')
graph.show()
```

#Sentiment Analysis of Netflix Content

```python
import pandas as pd
from textblob import TextBlob

# Initialize df3 from df with relevant columns
df3 = df[['Release_Date', 'Description']].copy()

# Process Release_Date to Release Year
df3['Release_Date'] = df3['Release_Date'].str.strip()
df3['Release Year'] = pd.to_datetime(df3['Release_Date'], format='%B %d, %Y', errors='coerce').dt.year

# Drop rows where 'Release Year' couldn't be parsed
df3 = df3.dropna(subset=['Release Year'])
df3['Release Year'] = df3['Release Year'].astype(int)

# Initialize 'Sentiment' column
df3['Sentiment'] = None

# Loop for sentiment analysis
for index, row in df3.iterrows():
    d = str(row['Description'])
    testimonial = TextBlob(d)
    p = testimonial.sentiment.polarity
    if p == 0:
        sent = 'Neutral'
    elif p > 0:
        sent = 'Positive'
    else:
        sent = 'Negative'
    df3.loc[index, 'Sentiment'] = sent

# Group by 'Release Year' and 'Sentiment'
df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Content')

# Filter for years >= 2015
df3 = df3[df3['Release Year'] >= 2014]

# Sort by 'Release Year'
df3 = df3.sort_values('Release Year')

print(df3)
barGraph = px.bar(df3, x="Release Year", y='Total Content' , color='Sentiment', title='Sentiment Analysis of Content on Netflix')
barGraph.show()
```
